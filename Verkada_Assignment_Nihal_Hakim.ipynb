```python
import pandas as  pd

# Read the csv files into Pandas dataframes

campaign_df = pd.read_csv('campaign.csv')
campaign_member_df = pd.read_csv('campaign_member.csv')
contact_df = pd.read_csv('contact.csv')
account_df = pd.read_csv('account.csv')
marketo_lead_df = pd.read_csv('marketo_lead.csv')
marketo_form_fill_df = pd.read_csv('marketo_form_fill.csv')
```

```python
# Count the occurrences of each contact_id and create a series of contact_ids that appear more than once
contact_counts = campaign_member_df['contact_id'].value_counts()
contact_ids_in_multiple_campaigns = contact_counts[contact_counts > 1]

# Display the contact_ids that are in multiple campaigns, (first column is contact_id, second column is how many campaigns they are in)
contact_ids_in_multiple_campaigns

# There are 17 contacts who are part of multiple campaigns, each contact in 2 campaigns
```

```python
# Create a list of affected contact ids (not a Series), so we can use it filter down the campaign member data
affected_contact_ids = contact_ids_in_multiple_campaigns.index.tolist()

# Filter down the campaign_member data to only show the rows with affected contacts
affected_campaign_members = campaign_member_df[campaign_member_df['contact_id'].isin(affected_contact_ids)]

# Display a list of affected campaign member IDs
affected_campaign_members['campaign_member_id'].tolist()
```

```python
# Sort the affected campaign members dataframe by contact ID

affected_campaign_members_sorted = affected_campaign_members.sort_values(['contact_id', 'created_date']).reset_index(drop=True)
affected_campaign_members_sorted

# we can see that for each contact, the campaign membership timestamps are within 2 hours of each other.
# we also can see that each contact was first added to 7012T00000KXg4XgDO then added to 7012T00000ZIfHxmNk
```

```python
# now lets cross-reference these prospects with all of their marketo form fills, using the marketo_lead table as an intermediary
# we do not want a frankenstein dataframe with duplicated data, so we will not merge/join data at this stage

# first we'll filter the marketo_lead table to only display leads that are on of the affected contacts
affected_marketo_leads = marketo_lead_df[marketo_lead_df['salesforce_contact_id'].isin(affected_contact_ids)]

# then we will filter the marketo_form_fill table to only display form fills where the lead is one of the affected
potentially_affected_form_fills = marketo_form_fill_df[marketo_form_fill_df['lead_id'].isin(affected_marketo_leads['marketo_lead_id'].tolist())]

# then lets sort it
potentially_affected_form_fills.sort_values(['lead_id', 'activity_date']).reset_index(drop=True)
```

```python
# now in order to visually make more sense of these form fills lets join contact_id and email to it
potentially_affected_form_fills_merged = pd.merge(potentially_affected_form_fills, marketo_lead_df[['marketo_lead_id','email','salesforce_contact_id']], left_on = 'lead_id', right_on = 'marketo_lead_id', how = 'inner').sort_values(['lead_id', 'activity_date']).reset_index(drop=True)
potentially_affected_form_fills_merged
```

```python
# lets introduce the account data into the mix
affected_campaign_members_merged_account = pd.merge(
  pd.merge(affected_campaign_members_sorted, contact_df[['contact_id','account_id']], left_on = 'contact_id', right_on = 'contact_id', how = 'inner'),
  account_df, left_on = 'account_id', right_on = 'account_id', how = 'inner'
)

affected_campaign_members_merged_account[affected_campaign_members_merged_account['campaign_id'] == '7012T00000ZIfHxmNk']

# there doesn't seem to be any anomolies with respect to account data merged with campaign member data
```

```python
# lets merge account data with form fills

potentially_affected_form_fills_merged_account = pd.merge(
  pd.merge(potentially_affected_form_fills_merged, contact_df[['contact_id','account_id']], left_on = 'salesforce_contact_id', right_on = 'contact_id', how = 'inner'),
  account_df, left_on = 'account_id', right_on = 'account_id', how = 'inner'
)

potentially_affected_form_fills_merged_account[potentially_affected_form_fills_merged_account['campaign_sfdc_id'] != '7012T00000ZIfHxmNk']

# there doesn't seem to be any anomolies with respect to account data merged with form fill data
```
